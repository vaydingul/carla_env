```python
SUMMARY_PATH = "/mnt/kuacc/scratch/users/vaydingul20/driving_model_data_20Hz_action_repeat_1/val/bev_world_occ_navigation_navigation_downsampled_ego_occ.npz"
```

```python
def open_summary(summary_path):
    data = np.load(summary_path, allow_pickle = True)
    return data["arr_0"].item()
```

```python
def plot_histogram(data, key):
    hist = data[key]["histogram"]
    bins = data[key]["bins"]
    plt.figure()
    plt.hist(bins[:-1], bins, weights = hist)
    plt.title(key)
```

```python
data = open_summary(SUMMARY_PATH)
for key in data.keys():
    plot_histogram(data, key)
```
